Expected Output#2
●	Side-by-side comparison and observations


```python
def is_palindrome(text):

  # Convert the text to lowercase and remove non-alphanumeric characters
  cleaned_text = ''.join(char.lower() for char in text if char.isalnum())

  # Check if the cleaned text is equal to its reverse
  return cleaned_text == cleaned_text[::-1]

# Now, let's test the function.
# Replace 'madam' with the string you want to check.
input_string = "madam"
if is_palindrome(input_string):
  print(f"'{input_string}' is a palindrome.")
else:
  print(f"'{input_string}' is not a palindrome.")

# You can try with another example:
input_string_2 = "hello"
if is_palindrome(input_string_2):
  print(f"'{input_string_2}' is a palindrome.")
else:
  print(f"'{input_string_2}' is not a palindrome.")
```
